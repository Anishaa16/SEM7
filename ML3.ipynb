```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to minimize
def function_to_minimize(x):
    return (x + 3) ** 2

def derivative_of_function(x):
    return 2 * (x + 3)

# Initial point
x = 2

# Hyperparameters
learning_rate = 0.1  # You can adjust this value based on your needs
iterations = 1000

# Lists to store the values for plotting
x_values = [x]
y_values = [function_to_minimize(x)]

# Gradient Descent algorithm
for i in range(iterations):
    gradient = derivative_of_function(x)
    x = x - learning_rate * gradient

    # Store the current values for plotting
    x_values.append(x)
    y_values.append(function_to_minimize(x))

# The final value of x should be close to the local minimum
min_x = x
min_y = function_to_minimize(min_x)

print(f"Local minimum found at x = {min_x}, y = {min_y}")

# Plot the function and the gradient descent process
x_range = np.linspace(min(x_values) - 1, max(x_values) + 1, 100)
y_range = [function_to_minimize(x) for x in x_range]

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='Function')
plt.scatter(x_values, y_values, color='red', marker='o', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()
```
